# School District Blood Lead Level Test Results

This data set analyzed the amount of lead absorption, in percentage, in various school districts in Pennsylvania. I ultimately concluded that the best neighborhood would have the school district with the lowest percentage of lead absorption. The metric I designed was lead_blood_levels, which measured the percentage of lead absorption in blood streams for every school district.


```python
import pandas as pd
import numpy as np
import geopandas
%matplotlib inline
import matplotlib.pyplot as plt
```

Next, I will import the data. Keep in mind, while I am importing all the data recorded from the years 2015-2020, I will only be analyzing the data from 2015. If needed, I will look at data from other years to compaare if one school district's percentages equal another.

```python
#importing the data
data = pd.read_csv("lead_blood_levels.csv")
data.head(10)
```

This data set was a bit tricky. For each school district there were percentages from the years 2015-2020. This meant that essentially each data “point” had more than one associated value. Instead of analyzing each individual school in each individual year I just decided to look at the results from 2015. For the first graph, I am choosing a scatter plot. I chose this because the percentages are not whole, easy numbers to read, so looking at the actual placement for a data point would be better represented than with a bar. The x axis represents the different school districts while the y axis represents the absorption amount in percent.


```python
#graph for 2015
data.sort_values(ascending=True, by="percentEBLL2015").plot(kind='scatter', x='SchoolDistrict', y ='percentEBLL2015', title ='COMPARING BLOOD ABSORPTION LEVELS IN SCHOOL DISTRICT IN 2015')
```

As seen above, the scatter plot is a bit hard to read. All of the school's names merged together. At least we can see where each

```python
endData = data.groupby("SchoolDistrict").sum()['percentEBLL2015']
endData.sort_values(ascending=True).head(10)
```

```python
endData.sort_values(ascending=False).head(10)
```

In conclusion, based on the scatter plot, the lowest lead absorption percentages in school districts in 2015, with values of 0%, were Allegheny Valley and Bethel Park. Since both of these districts have the same value, I decided to just take a look at both of them in the next two years. In 2016 they also both had 0% lead absorption so I took a look at 2017. In 2017, Allegheny Valley had 0.92% absorption while Bethel Park had 0.61% absorption. Therefore, after analyzing and compiling all of this information, Bethel Park is the best School District and Neighborhood to live in.
